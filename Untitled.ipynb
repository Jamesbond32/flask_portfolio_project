```python
with open('requirements.txt', 'r+') as file:
    file_list = file.readlines()
```

```python
file_list
```

```python
new_list = []

for line in file_list:
    if '=' in line:
        
        split_list = str(line).split('=', maxsplit=1)
        print(split_list[0] +' == ', split_list[1])
        new_list.append((split_list[0] +' == ', split_list[1]))
```

```python
string_list = []

for item in new_list:
    print(str(item[0] + str(item[1])))
    string_list.append(str(item[0] + str(item[1])))
    
```

```python
string_list
```

```python
req_list = []
```

```python
string_list[0]
```

```python
for item in string_list:
    if '=' in item:
        start = item.find('=')+2
        index = item.find('=', start)
        
        print(item[0:index])
        req_list.append(item[0:index])
```
